#### Importando as bibliotecas necessárias

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Lendo os dados

```python
df = pd.read_csv('medical_examination.csv')
df.head()
```

#### Redefinindo unidades de medida 

Idade(dias) -> idade(anos)

Altura(cm) -> altura(metros)

```python
df['age'] = df['age'] // 365
```

```python
df['height'] = df['height'] / 100
```

#### Definindo uma coluna 'sobrepeso' através do calculo do IMC

```python
df['overweight'] = np.where(df['weight'] / np.square(df['height']) > 25, 1, 0)
df.head()
```

#### Normalizando as colunas 'colesterol' e 'glicose'

```python
df['cholesterol'] = np.where(df['cholesterol'] == 1, 0, 1)
df.head()
```

```python
df['gluc'] = np.where(df['gluc'] == 1, 0, 1)
df.head()
```

#### Definindo o DataFrame para o Categorical Plot

```python
df_cat = ['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight']
df_cat = pd.melt(df, id_vars = ['cardio'], value_vars = df_cat)
```

```python
df_cat.head()
```

```python
df_cat = pd.DataFrame(df_cat.groupby(['cardio','variable','value'])['value'].count()).rename(columns={'value':'total'}).reset_index()
df_cat.head()
```

#### Catplot

```python
sns.catplot(data = df_cat, x = 'variable', y = 'total', hue = 'value', col = 'cardio', kind = 'bar', palette = 'husl')
```

#### Limpando os dados

Temos dados incorretos quando:

- A pressão diastólica é maior que a sistólica 
- Altura é menor que o 2,5th percentil 
- Altura é maior que o 97,5th percentil 
- O peso é menor que o 2,5th percentil
- O peso é maior que 97,5th percentil 

```python
df_heat = df[
        (df['ap_lo'] <= df['ap_hi']) &
        (df['height'] >= df['height'].quantile(0.025)) &
        (df['height'] <= df['height'].quantile(0.975)) &
        (df['weight'] >= df['weight'].quantile(0.025)) &
        (df['weight'] <= df['weight'].quantile(0.975))
]
```

```python
df_heat
```

#### Calculando a matriz de correlação

```python
matrix_corr = df_heat.corr()
```

```python
matrix_corr
```

```python
matrix_corr.describe()
```

#### Filtrando apenas o triângulo superior da matriz

```python
mask = np.triu(matrix_corr)
```

#### Setando o tamanho da figura e plotando o Heatmap

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.heatmap(matrix_corr, linewidths=0.8, mask=mask, vmax=0.6, cmap='Reds')
```
